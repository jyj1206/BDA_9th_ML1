# 교차검증

## cross_val_score

```python
# 라이브러리 불러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer
import numpy as np
```

```python
# 유방암 데이터 불러오기
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
```

```python
# 모델 불러오기
clf = DecisionTreeClassifier(random_state=42)
```

```python
clf
```

```python
# cross_val_score 실행
from sklearn.model_selection import cross_val_score, StratifiedKFold
stkfold= StratifiedKFold(n_splits=5) # StratifedKFold 객체 생성
scores = cross_val_score(clf, X, y, cv=stkfold, scoring='accuracy')
```

```python
# 결과 확인
scores.mean()
```

## cross_validate

```python
# cross_validate 실행
from sklearn.model_selection import cross_validate
result = cross_validate(clf, X, y, cv=stkfold, scoring='accuracy', return_train_score=True)
```

```python
result
```

```python
scorings = ['accuracy','precision','recall']
cross_validate(clf, X,y, cv =stkfold, scoring=  scorings, return_train_score=True)
```

```python
# 펭귄 데이터 교차검증 실행
import seaborn as sns
df = sns.load_dataset('penguins')

# 결측치 처리(0으로 채움)
df.fillna(0, inplace=True)
```

```python
# 클래스 종류 확인
df.species.value_counts()
```

```python
df
```

```python
# X, y 데이터 준비
df_x = df.drop(['island','species', 'sex'], axis=1)
df_y = df.species
```

```python
clf = RandomForestClassifier(random_state=42)
```

```python
# 교차검증 (오류)
scorings = ['accuracy', 'precision', 'recall', 'f1']
cross_validate(clf, df_x, df_y, cv=stkfold, scoring=scorings, return_train_score=True)
```

> ValueError: Target is multiclass but average='binary'. Please choose another average setting, one of [None, 'micro', 'macro', 'weighted'].
> - 위와같은 경고 발생, 이 경우 오류에 해결방법 처럼 기본적인 'precision', 'recall', 'f1'은 average 파라미터를 'binary'로 사용 때문에 [None, 'micro', 'macro', 'weighted'] 이중 하나를 선택해주어야함

```python
# 교차검증 (micro)
scorings = ['accuracy', 'precision_micro', 'recall_micro', 'f1_micro']
cross_validate(clf, df_x, df_y, cv=stkfold, scoring=scorings, return_train_score=True)
```

```python
## 다양한 모델을 사용
#DecisionTreeClassifier, RandomForestClassifier, SVC

models = [DecisionTreeClassifier(random_state=42), RandomForestClassifier(random_state=42), SVC(random_state=42)]
name =['DT','RF','LR']

for model, name in zip(models, name):
    print('###사용 알고리즘',name,'###')
    for score in ['accuracy','precision_micro','recall_micro','f1_micro']:
        print(score)
        print("-" * 20)
        print(cross_val_score(model, df_x,df_y, cv=stkfold, scoring=score))
```

# 임계점

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
# train, test 분리
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3,random_state=42)
```

```python
# 모델 생성
clf = RandomForestClassifier(random_state=42)
```

```python
def get_clf_eval(y_test, pred):
    conf = confusion_matrix(y_test, pred)
    acc = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred, average='micro')
    recall = recall_score(y_test, pred, average ='micro')

    print(conf)
    print(acc, precision, recall)
```

```python
# 모델 학습
clf.fit(X_train, y_train)

# 예측
pred = clf.predict(X_test)

# 성능 평가
get_clf_eval(y_test, pred)
```

```python
# 각 클래스에 속할 확률을 예측
clf.predict_proba(X_test)
```
